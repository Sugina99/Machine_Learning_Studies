```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression,Ridge,Lasso
from sklearn.model_selection import train_test_split
```

```python
# load data
data=pd.read_csv("Melbourne_housing_FULL.csv")
```

```python
data.head()
```

```python
data.shape
```

```python
# do eda later..just testing lasso and ridge regression..so taking few cols only
```

```python
data.nunique()
```

```python
# lets use limited columns which makes more sence for serving our purpose
cols_to_use=['Suburb','Rooms','Type','Method','SellerG','Regionname','Propertycount','Distance','CouncilArea','Bedroom2','Bathroom','Car','Landsize','Price','BuildingArea']
```

```python
data=data[cols_to_use]
```

```python
data.head()
```

```python
data.shape
```

```python
data.isna().sum()
```

```python
# fill missing value with mode and median
data['Regionname']=data['Regionname'].fillna(data['Regionname'].mode()[0])
data['CouncilArea']=data['CouncilArea'].fillna(data['CouncilArea'].mode()[0])

data['Propertycount']=data['Propertycount'].fillna(data['Propertycount'].median())
data['BuildingArea']=data['BuildingArea'].fillna(data['BuildingArea'].median())

data['Distance']=data['Distance'].fillna(data['Distance'].median())
data['Bedroom2']=data['Bedroom2'].fillna(data['Bedroom2'].mode()[0])
data['Bathroom']=data['Bathroom'].fillna(data['Bathroom'].mode()[0])
data['Car']=data['Car'].fillna(data['Car'].mode()[0])
data['Landsize']=data['Landsize'].fillna(data['Landsize'].median())
```

```python
data.isna().sum()
```

```python
# dependend varible 'Price' has missing values there fore we removing the column.
data.dropna(inplace=True)
```

```python
data.shape
```

```python
data.head()
```

```python
# encoding :- onehot encoding
data=pd.get_dummies(data,drop_first=True)
data.head()
```

```python
X=data.drop('Price',axis=1)
y=data['Price']
```

```python
X
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=21)
```

```python
scaler=StandardScaler()
scaler.fit_transform(X_train)
```

```python
actual_df=pd.DataFrame(scaler.fit_transform(X_train))
```

```python
actual_df.columns=X.columns
```

```python
actual_df
```

```python
##### using Ridge Regression
lr=LinearRegression()
lr.fit(X_train,y_train)
```

```python
lr.predict(X_test)
```

```python
train_accuracy=lr.score(X_train,y_train)
```

```python
test_accuracy=lr.score(X_test,y_test)
```

```python
print(f"train accuracy is: {train_accuracy}")
print(f"test accuracy is: {test_accuracy}")
```

```python
# alpha=50
ridge_reg=Ridge(alpha=50) #alpha is lambda value
ridge_reg.fit(X_train,y_train)
```

```python
ridge_reg.predict(X_test)
```

```python
ridge_reg.score(X_train,y_train)
```

```python
ridge_reg.score(X_test,y_test)
```

```python
# alpha=10
ridge_reg=Ridge(alpha=10) #alpha is lambda value
ridge_reg.fit(X_train,y_train)
```

```python
ridge_reg.score(X_train,y_train)
```

```python
ridge_reg.score(X_test,y_test)
```

```python
ridge_reg=Ridge(alpha=0.1) #alpha is lambda value
ridge_reg.fit(X_train,y_train)
train_accuracy=ridge_reg.score(X_train,y_train)
test_accuracy=ridge_reg.score(X_test,y_test)
print(f"train_accuracy = {train_accuracy}")
print(f"test_accuracy = {test_accuracy}")
```

```python
##### using LASSO Regression
lasso_reg=Lasso(alpha=50)
lasso_reg.fit(X_train,y_train)
```

```python
train_accuracy=lasso_reg.score(X_train,y_train)
test_accuracy=lasso_reg.score(X_test,y_test)
print(f"train_accuracy = {train_accuracy}")
print(f"test_accuracy = {test_accuracy}")
```

```python
lasso_reg=Lasso(alpha=0.1)
lasso_reg.fit(X_train,y_train)
train_accuracy=lasso_reg.score(X_train,y_train)
test_accuracy=lasso_reg.score(X_test,y_test)
print(f"train_accuracy = {train_accuracy}")
print(f"test_accuracy = {test_accuracy}")
```
