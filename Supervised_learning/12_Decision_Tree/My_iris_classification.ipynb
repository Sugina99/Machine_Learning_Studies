```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings("ignore") 
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import mean_squared_error
from sklearn.datasets import load_iris
```

```python
# loading data set
iris=load_iris()
iris_df=pd.DataFrame(iris.data,columns=iris.feature_names)
iris_df['Species']=iris.target
iris_df.head()
```

```python
# X and y values
X=iris_df.drop(['Species'],axis=1)
y=iris_df['Species']
```

```python
#  splitting
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=17)
```

```python
# dimension
print(f"diamension of X_train is {X_train.shape}")
print(f"diamension of X_test is {X_test.shape}")
```

```python
dt_classifier=DecisionTreeClassifier()
dt_classifier.fit(X_train,y_train)
```

```python
score=dt_classifier.score(X_train,y_train)
print(f"train score is {score}")
score=dt_classifier.score(X_test,y_test)
print(f"test score is {score}")
y_pred=dt_classifier.predict(X_test)
```

```python
mse= mean_squared_error(y_test,y_pred)
print(f"mse {mse}")
rmse=np.sqrt(mse)
print(f"rmse {rmse}")
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(dt_classifier,filled=True)
```

```python
dt_classifier=DecisionTreeClassifier(criterion="entropy",max_depth=5)
dt_classifier.fit(X_train,y_train)

score=dt_classifier.score(X_train,y_train)
print(f"train score is {score}")
score=dt_classifier.score(X_test,y_test)
print(f"test score is {score}")
y_pred=dt_classifier.predict(X_test)

mse= mean_squared_error(y_test,y_pred)
print(f"mse {mse}")
rmse=np.sqrt(mse)
print(f"rmse {rmse}")
```
