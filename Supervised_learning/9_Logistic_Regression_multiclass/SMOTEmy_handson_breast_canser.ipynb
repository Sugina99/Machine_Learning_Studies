```python
data.drop(['id','Unnamed: 32'],axis=1,inplace=True)
```

```python
# one hot encoding
data=pd.get_dummies(data,columns=['diagnosis'],prefix='daignosis',drop_first=True)
#  or data['diagnosis']=data['diagnosis'].map({'B':0,'M':1})
```

```python
X=data.drop('daignosis_M',axis=1)
y=data['daignosis_M']
```

```python
y
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
X_train
```

```python
classifier=LogisticRegression()
classifier.fit(X_train,y_train)
```

```python
classifier.score(X_train,y_train)
```

```python
classifier.score(X_test,y_test)
```

```python
from imblearn.over_sampling import SMOTE
```

```python
sm=SMOTE(random_state=42)
```

```python
X_res,y_res=sm.fit_resample(X_train,y_train)
```

```python
y_train.value_counts()
```

```python
y_res.value_counts()
```

```python
log_rig=LogisticRegression()
log_rig.fit(X_res,y_res)
```

```python
log_rig.score(X_res,y_res)
```

```python
log_rig.score(X_test,y_test)
```
